```python


SDURL_BUAS = "http://194.171.191.226:3061/genaichallenge"
SDURL_BUAS_TXT2IMG = f"{SDURL_BUAS}/sdapi/v1/txt2img"

def generate_image_buas(prompt, negative_prompt, filename):
  payload = {
    "prompt": prompt,
    "steps": 6,
    "cfg_scale": 2,
    "randomize_seed": -1,
    "seed": -1,
    "enable_hr": False,
    "denoising_strength": 0,
    "firstphase_width": 0,
    "firstphase_height": 0,
    "hr_scale": 2,
    "hr_upscaler": "Hr Upscaler",
    "hr_second_pass_steps": 0,
    "hr_resize_x": 0,
    "hr_resize_y": 0,
    "styles": [ ],
    "subseed": -1,
    "subseed_strength": 0,
    "seed_resize_from_h": -1,
    "seed_resize_from_w": -1,
    "sampler_name": "DPM++ SDE",
    "batch_size": 1,
    "n_iter": 1,
    "width": 512,
    "height": 512,
    "restore_faces": False,
    "tiling": False,    
    "negative_prompt": negative_prompt,
    "eta": 0,
    "s_churn": 0,
    "s_tmax": 0,
    "s_tmin": 0,
    "s_noise": 1,
    "override_settings": {},
    "override_settings_restore_afterwards": True,
    "script_args": []
  }  
  response = requests.post(SDURL_BUAS_TXT2IMG, json=payload)  
  if response.ok:
    imgdata = response.json()['images'][0]
    bytestream = io.BytesIO(base64.b64decode(imgdata.split(",",1)[0]))    
    with open(filename, "wb") as outfile:
      outfile.write(bytestream.getbuffer())
    print("Image generation succeeded: " + filename)
  else:
    print("Image generation failed!")

num_images = 50
prompt = "picture of a surprised human facial expression"
for i in range(num_images):
  generate_image_buas(prompt, "", prompt + str(i) + ".png")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
import time

def download_image(image_url, directory, index):
    """
    Downloads an image from a given URL into a specified directory.
    """
    try:
        response = requests.get(image_url)
        if response.status_code == 200:
            file_path = os.path.join(directory, f'image_{index}.jpg')
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded {file_path}")
    except Exception as e:
        print(f"Error downloading {image_url}: {e}")

def scrape_images(search_term, limit):
    """
    Scrapes images from DuckDuckGo based on a search term up to a specified limit.
    """
    # Initialize the Chrome driver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    # Go to DuckDuckGo's image search page
    driver.get(f"https://duckduckgo.com/?q={search_term}&iax=images&ia=images")
    time.sleep(3)  # Allow time for the page to load

    # Scroll the page to ensure that images are loaded
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for new images to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # Find image elements on the page
    images = driver.find_elements(By.CSS_SELECTOR, "img.tile--img__img.js-lazyload")
    images = images[:limit]  # Limit the number of images to download

    # Create a directory for the downloaded images
    directory = f"downloaded_images_{search_term.replace(' ', '_')}"
    if not os.path.exists(directory):
        os.makedirs(directory)

    # Download the images
    for index, image in enumerate(images):
        image_url = image.get_attribute('src')
        if image_url:
            download_image(image_url, directory, index)

    driver.quit()
    print(f"Finished downloading images to {directory}.")

# Example usage

search_term = "scared expression"  # Specify the search term here
limit = 200  # Specify the maximum number of images to download

scrape_images(search_term, limit)
#emotions = ['happy', 'angry', 'disgusted', 'scared', 'neutral', 'sad']

#for emotion in emotions :
#    scrape_images(f"{emotion} person face expression",200)
#    scrape_images(f"{emotion} black personface expression",20)
```

```python
import os

def rename_files(folder_path, new_name_prefix):
    # Get the list of files in the folder
    files = os.listdir(folder_path)
    
    # Iterate through each file
    for index, file_name in enumerate(files):
        # Generate the new file name
        new_file_name = f"{new_name_prefix}_{index + 1}" + os.path.splitext(file_name)[1]
        
        # Get the current file path and the new file path
        current_file_path = os.path.join(folder_path, file_name)
        new_file_path = os.path.join(folder_path, new_file_name)
        
        # Rename the file
        os.rename(current_file_path, new_file_path)
        print(f"Renamed {file_name} to {new_file_name}")

# Example usage:
folder_path = r"C:\Users\zakar\Desktop\School\BLOCK_C\WEEK 9 10\emotion\Fear"
new_name_prefix = "fear"
rename_files(folder_path, new_name_prefix)
```
