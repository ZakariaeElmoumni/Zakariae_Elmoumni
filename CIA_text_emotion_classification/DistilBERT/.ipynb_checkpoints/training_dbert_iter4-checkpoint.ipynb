```python
### 1. IMPORTS ###
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from transformers import DistilBertTokenizer, TFDistilBertForSequenceClassification
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import confusion_matrix
import seaborn as sns
import random

# Additional imports for data augmentation
import nlpaug.augmenter.word as naw
import nlpaug.augmenter.char as nac
import nlpaug.augmenter.sentence as nas
import nltk
from nltk.corpus import wordnet

# Download required NLTK resources
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('omw-1.4', quiet=True)
```

```python
### 2. LOADING DATASETS ###
df_main = pd.read_csv('../datasets/dataset_v2.csv', encoding="latin1")
df_test = pd.read_csv('../datasets/group 11_url1.csv', encoding="latin1")

df_main = df_main.dropna(subset=["text", "dominant_emotion"])
df_main = df_main.rename(columns={"dominant_emotion": "emotion"})

df_test = df_test.dropna(subset=["Translation", "Emotion"])
df_test = df_test.rename(columns={"Translation": "text", "Emotion": "emotion"})

### 3. MAPPING OF THE TEST SET ###
# Mapping emotions
df_test['emotion'] = df_test['emotion'].map({
    'neutral': 'neutral', 'disgust': 'disgust', 'surprise': 'surprise', 'curiosity': 'surprise',
    'excitement': 'happiness', 'fear': 'fear', 'optimism': 'happiness', 'nervousness': 'fear',
    'confusion': 'neutral', 'annoyance': 'anger', 'caring': 'happiness', 'disapproval': 'anger',
    'approval': 'happiness', 'disappointment': 'sadness', 'admiration': 'happiness',
    'desire': 'happiness', 'amusement': 'happiness', 'sadness': 'sadness', 'pride': 'happiness',
    'gratitude': 'happiness', 'realization': 'happiness', 'relief': 'happiness',
    'remorse': 'sadness', 'joy': 'happiness', 'love': 'happiness', 'anger': 'anger'
}).fillna('Other')

# Override with 'neutral' if intensity is mild
df_test.loc[df_test['Intensity'].str.lower() == 'mild', 'emotion'] = 'neutral'
```

```python
### 5. SIMPLE BUT ROBUST TEXT AUGMENTATION ###

# This part of the code really made the difference for the new iteration since its giving more examples of other emotions
def get_synonyms(word):
    """Get synonyms for a word using WordNet."""
    synonyms = []
    try:
        for syn in wordnet.synsets(word):
            for lemma in syn.lemmas():
                synonym = lemma.name().replace('_', ' ')
                if synonym != word and synonym not in synonyms:
                    synonyms.append(synonym)
        return synonyms
    except:
        return []

def simple_synonym_replacement(text, n=1):
    """Replace n words in the text with their synonyms."""
    words = text.split()
    if len(words) <= 3:  # Skip very short texts
        return text
        
    new_words = words.copy()
    random_word_indices = random.sample(range(len(words)), min(n, len(words)))
    
    for idx in random_word_indices:
        word = words[idx]
        # Skip short words, punctuation, etc.
        if len(word) <= 3 or not word.isalpha():
            continue
            
        synonyms = get_synonyms(word)
        if synonyms:
            new_words[idx] = random.choice(synonyms)
    
    return ' '.join(new_words)

def simple_random_swap(text, n=1):
    """Randomly swap n pairs of words in the text."""
    words = text.split()
    if len(words) <= 3:  # Skip very short texts
        return text
        
    new_words = words.copy()
    for _ in range(min(n, len(words)//2)):
        idx1, idx2 = random.sample(range(len(new_words)), 2)
        new_words[idx1], new_words[idx2] = new_words[idx2], new_words[idx1]
    
    return ' '.join(new_words)

def simple_random_deletion(text, p=0.1):
    """Randomly delete words from the text with probability p."""
    words = text.split()
    if len(words) <= 3:  # Skip very short texts
        return text
        
    # Make sure we don't delete too many words
    p = min(p, 0.3)
    
    new_words = []
    for word in words:
        if random.random() >= p:
            new_words.append(word)
    
    # If we deleted all words, return the original text
    if not new_words:
        return text
    
    return ' '.join(new_words)

def robust_text_augmentation(df, target_column='emotion', text_column='text', 
                            minority_classes=None, target_count=None):
    """
    Perform simple but robust text augmentation for minority classes.
    """
    df_aug = df.copy()
    
    # If minority_classes is not specified, determine them automatically
    if minority_classes is None:
        class_counts = df[target_column].value_counts()
        median_count = class_counts.median()
        minority_classes = class_counts[class_counts < median_count].index.tolist()
    
    # If target_count is not specified, set it to median class count
    if target_count is None:
        target_count = int(df[target_column].value_counts().median())
    
    print(f"Augmenting minority classes: {minority_classes}")
    print(f"Target count per class: {target_count}")
    
    augmented_samples = []
    
    # Process each minority class
    for emotion in tqdm(minority_classes, desc="Augmenting classes"):
        minority_samples = df[df[target_column] == emotion]
        samples_needed = target_count - len(minority_samples)
        
        if samples_needed <= 0:
            continue
            
        # Determine augmentation methods based on sample needs
        augmentation_methods = [
            lambda x: simple_synonym_replacement(x, n=1),
            lambda x: simple_synonym_replacement(x, n=2),
            lambda x: simple_random_swap(x, n=1),
            lambda x: simple_random_deletion(x, p=0.1),
            lambda x: x  # Original text as a fallback
        ]
        
        # Create augmented samples through multiple passes
        samples_created = 0
        max_iterations = 3  # Limit number of passes through the dataset
        
        for iteration in range(max_iterations):
            if samples_created >= samples_needed:
                break
                
            # Go through each sample in the minority class
            for _, row in tqdm(minority_samples.iterrows(), 
                              desc=f"Augmenting {emotion} (iteration {iteration+1})",
                              total=len(minority_samples)):
                
                if samples_created >= samples_needed:
                    break
                    
                text = row[text_column]
                
                # Skip texts that are too short
                if len(text.split()) <= 3:
                    continue
                
                # Try different augmentation methods
                for aug_method in augmentation_methods:
                    try:
                        aug_text = aug_method(text)
                        
                        # Skip if augmentation didn't change the text
                        if aug_text == text:
                            continue
                            
                        # Create new row with augmented text
                        new_row = row.copy()
                        new_row[text_column] = aug_text
                        augmented_samples.append(new_row)
                        samples_created += 1
                        
                        if samples_created >= samples_needed:
                            break
                    except Exception as e:
                        continue
        
        # If we still need more samples, duplicate some existing ones
        if samples_created < samples_needed:
            additional_samples_needed = samples_needed - samples_created
            print(f"Could only generate {samples_created} unique augmented samples for {emotion}.")
            print(f"Adding {additional_samples_needed} duplicated samples to reach target count.")
            
            # Get all samples for this class (original + augmented)
            all_class_samples = list(minority_samples.index)
            all_class_samples.extend(range(len(df), len(df) + samples_created))
            
            # Randomly select samples to duplicate
            if all_class_samples:
                duplicate_indices = random.choices(all_class_samples, k=additional_samples_needed)
                
                for idx in duplicate_indices:
                    if isinstance(idx, int) and idx >= len(df):
                        # This is an already augmented sample
                        aug_idx = idx - len(df)
                        if aug_idx < len(augmented_samples):
                            augmented_samples.append(augmented_samples[aug_idx].copy())
                    else:
                        # This is an original sample
                        augmented_samples.append(df.loc[idx].copy())
    
    # Create DataFrame with augmented samples
    if augmented_samples:
        df_augmented = pd.DataFrame(augmented_samples)
        df_combined = pd.concat([df_aug, df_augmented], ignore_index=True)
        
        print(f"Added {len(df_augmented)} augmented samples")
        print(f"New class distribution:")
        print(df_combined[target_column].value_counts())
        
        return df_combined
    else:
        print("No augmented samples were generated")
        return df_aug
```

```python
### 6. HANDLING CLASS IMBALANCE WITH AUGMENTATION ###
# Define minority classes
minority_classes = ['disgust']  # We'll focus on the most imbalanced class

# First, augment the minority classes
df_augmented = robust_text_augmentation(
    df_main,
    target_column='emotion',
    text_column='text',
    minority_classes=minority_classes,
    target_count=15000  # Target count closer to other classes
)

# After augmentation, apply more balanced resampling for final balancing
emotions = df_augmented['emotion'].value_counts()
target_count = 20000  # Target count for all classes

balanced_dfs = []
for emotion, count in emotions.items():
    subset = df_augmented[df_augmented['emotion'] == emotion]
    if count > target_count:
        # Downsample majority classes
        balanced_dfs.append(resample(subset, replace=False, n_samples=target_count, random_state=42))
    elif count < target_count:
        # Upsample minority classes
        balanced_dfs.append(resample(subset, replace=True, n_samples=target_count, random_state=42))
    else:
        balanced_dfs.append(subset)

df_balanced = pd.concat(balanced_dfs)
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print("\nFinal balanced dataset emotion distribution:")
print(df_balanced['emotion'].value_counts())
```

```python
### 7. SPLITING INTO TRAIN VALIDATION SETS ###
# 85% train, 15% test
train_df, val_df = train_test_split(df_balanced, test_size=0.15, random_state=42, stratify=df_balanced['emotion'])


### 8. ENCODING THE LABELS ###
label_encoder = LabelEncoder()
train_df['label'] = label_encoder.fit_transform(train_df['emotion'])
val_df['label'] = label_encoder.transform(val_df['emotion'])
df_test['label'] = label_encoder.transform(df_test['emotion'])

### 9. TOKENIZING THE INPUTS ###
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize_texts(texts):
    return tokenizer(list(texts), truncation=True, padding=True, return_tensors="tf", max_length=128)

train_encodings = tokenize_texts(train_df["text"])
val_encodings = tokenize_texts(val_df["text"])
test_encodings = tokenize_texts(df_test["text"])

train_labels = tf.convert_to_tensor(train_df["label"].values, dtype=tf.int32)
val_labels = tf.convert_to_tensor(val_df["label"].values, dtype=tf.int32)
test_labels = tf.convert_to_tensor(df_test["label"].values, dtype=tf.int32)
```

```python
### 10. FINAL DATASET PREPARATION ###
# Putting labels and text together for batching
def create_dataset(encodings, labels, batch_size=16, is_training=False):
    dataset = tf.data.Dataset.from_tensor_slices((dict(encodings), labels))
    if is_training:
        dataset = dataset.shuffle(1000)
    return dataset.batch(batch_size)

train_dataset = create_dataset(train_encodings, train_labels, is_training=True)
val_dataset = create_dataset(val_encodings, val_labels)
test_dataset = create_dataset(test_encodings, test_labels)

### 11. MODEL SETUP ###
# Using larger dropout and weight decay to prevent overfitting on augmented data
model = TFDistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased", 
    num_labels=len(label_encoder.classes_),
    dropout=0.4  # Increased dropout for regularization
)

# Learning rate schedule with warmup and decay
initial_learning_rate = 3e-5  # Slightly lower learning rate for augmented data
lr_schedule = tf.keras.optimizers.schedules.PolynomialDecay(
    initial_learning_rate,
    decay_steps=train_dataset.cardinality().numpy() * 5,  # 5 epochs
    end_learning_rate=1e-6,
    power=1.0  # Linear decay
)

optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule, weight_decay=0.025)
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metrics = [tf.keras.metrics.SparseCategoricalAccuracy()]

model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)

### 12. TRAINING ###
# Early stopping with increased patience for augmented data
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=2,  # Increased patience
    restore_best_weights=True,
    verbose=1
)

# Class weights to further balance training (focus more on minority classes)
class_weights = {}
for i, emotion in enumerate(label_encoder.classes_):
    # Calculate inverse frequency as weight
    count = df_balanced['emotion'].value_counts()[emotion]
    class_weights[i] = 1.0  # Equal weighting since we've balanced the dataset

# Training
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=12,  # Train for more epochs with augmented data
    callbacks=[early_stopping],
    class_weight=class_weights
)

model.save_weights('./dbert_iter4_weights.h5')
```

```python
### 13. PLOTTING THE LEARNING CURVES ###
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.title('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['sparse_categorical_accuracy'], label='train')
plt.plot(history.history['val_sparse_categorical_accuracy'], label='validation')
plt.title('Accuracy')
plt.legend()
plt.savefig('learning_curves_augmented.png')

### 14. TESTING ###
print("\nEvaluating on validation set...")
val_results = model.evaluate(val_dataset)
print("Validation Loss:", val_results[0])
print("Validation Accuracy:", val_results[1])

print("\nEvaluating on test set...")
test_results = model.evaluate(test_dataset)
print("Test Loss:", test_results[0])
print("Test Accuracy:", test_results[1])

predictions = model.predict(test_dataset).logits
predicted_labels = np.argmax(predictions, axis=-1)

# Print classification report
print("\nClassification Report:")
print(classification_report(test_labels, predicted_labels, target_names=label_encoder.classes_))

### 15. ERROR ANALYSIS ###
test_texts_list = list(df_test["text"])
test_emotions = list(df_test["emotion"])

# Get indices of misclassified examples
misclassified_indices = np.where(predicted_labels != test_labels)[0]

print("\nSample of misclassified examples:")
for i in misclassified_indices[:20]:  # Show first 20 errors
    print(f"Text: {test_texts_list[i][:100]}...")
    print(f"True emotion: {test_emotions[i]}")
    print(f"Predicted emotion: {label_encoder.classes_[predicted_labels[i]]}")
    print("-" * 50)
```
